```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the two csv files 
student_path = "Resources/students_complete.csv"
school_path = "Resources/schools_complete.csv"

# Readign the files and storing them into Pandas Dataframes
student_data_df = pd.read_csv(student_path)
school_data_df = pd.read_csv(school_path)

# combining the data into one dataset
combined_df = pd.merge(student_data_df, school_data_df, on = "school_name", how = "outer")
```

```python
combined_df
```

```python
# District summary 
```

```python
# Total Schools
schools = combined_df["school_name"].unique()
total_schools = len(schools)

# Total Students
total_students = combined_df["student_name"].count()

# Total Budget
# total_budget = combined_df["budget"].sum()
# this adds total budget per student, because thats the number of the table right?
# but I would only want to add one budget per school
# so maybe use school df
total_budget = school_data_df["budget"].sum()
#total_budget

# Average Math Score
average_math_score = combined_df["math_score"].mean()
#average_math_score

# Average Reading Score
average_reading_score = combined_df["reading_score"].mean()
#average_reading_score

# % Passing Math (The percentage of students that passed math.)
# counting 70 or over as passing
# the conditional statement comes back as a list of true and false, so need to get the length 
# len out the front counts everything, bothtrue and false
# how do I count just true
# setting it with math_score as the column has made it just a series
# so can do .count()
# still need to round everything later
number_passed_math = combined_df.loc[(combined_df["math_score"] >= 70), "math_score"].count()
percentage_passed_math = (number_passed_math/total_students)*100
percentage_passed_math
#number_passed_math

# % Passing Reading (The percentage of students that passed reading.)
number_passed_reading = combined_df.loc[(combined_df["reading_score"] >= 70), "reading_score"].count()
percentage_passed_reading = (number_passed_reading/total_students)*100
percentage_passed_reading

# % Overall Passing (The percentage of students that passed math and reading.)
number_passed_overall = combined_df.loc[(combined_df["reading_score"] >= 70) & (combined_df["math_score"] >= 70), "student_name"].count()
percentage_passed_overall = (number_passed_overall/total_students)*100
percentage_passed_overall


```

```python
# ccreating dataframe for info above
district_summary_df = pd.DataFrame([{"Total Schools":total_schools, 
                                    "Total Students":total_students, 
                                    "Total Budget":total_budget, 
                                    "Average Math Score":average_math_score, 
                                    "Average Reading Score":average_reading_score, 
                                    "% Passing Math":percentage_passed_math, 
                                    "% Passing Reading":percentage_passed_reading, 
                                    "% Overall Passing": percentage_passed_overall}])
district_summary_df
```
